```python
import numpy as nm
import pandas as pd
import matplotlib.pyplot as mtp  
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("ndata.csv")
df = df.drop(['Unnamed: 32','id'], axis=1)
df.diagnosis.replace(['M', 'B'], [1, 0], inplace=True)
df.head()
```

#### Finding the Skewness

```python
df_skew = pd.DataFrame(data={
    'skewness': df.skew()
})

df_skew
```

#### Log transformation our skewed data

```python
df.radius_se = np.log(df.radius_se + 1)
df.perimeter_se = np.log(df.perimeter_se + 1)
df.area_se = np.log(df.area_se)
df.concavity_se = np.log(df.concavity_se + 1)
df.fractal_dimension_se = np.log(df.fractal_dimension_se)
```

#### Detect and remove the outliers with Z-score

```python
from scipy import stats
z = np.abs(stats.zscore(df))
print(df.shape)
```

```python
df = df[(z < 3).all(axis=1)]
print(df.shape)
df.head()
```

Note:

1.0 (Orange) = Benign (No Cancer)

0.0 (Blue) = Malignant (Cancer)

```python
df['diagnosis'].value_counts()
```

```python
sns.countplot(df['diagnosis'], label = "Count") 
```

```python
# We drop our "Diagnosis" feature and use all the remaining features in our dataframe to train the model.
X = df.drop(['diagnosis'], axis = 1)
X.head()
```

```python
y = df['diagnosis']
y.head()
y.shape
```

#### Feature Engineering

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 30)
```

```python
print ('The size of our training "X" (input features) is', X_train.shape)
print ('\n')
print ('The size of our testing "X" (input features) is', X_test.shape)
print ('\n')
print ('The size of our training "y" (output feature) is', y_train.shape)
print ('\n')
print ('The size of our testing "y" (output features) is', y_test.shape)
```

#### Linear Model

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='linear')
svc_model.fit(X_train, y_train)
y_predict = svc_model.predict(X_test)
```

```python
# Import metric libraries
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = nm.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```

```python
sns.heatmap(confusion, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the Linear Model:",accuracy_score(y_test,y_predict))
```

#### Polynomial kernel

```python
from sklearn.svm import SVC
svc_model = SVC(kernel='poly', degree=8)
svc_model.fit(X_train, y_train)
```

```python
y_predict = svc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = nm.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```

```python
sns.heatmap(confusion, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
print("Accuracy of the Polynomial Model:",accuracy_score(y_test,y_predict))
```
